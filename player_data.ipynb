```
def pre_data(player_id, year):
    url = "http://stats.espncricinfo.com/ci/engine/player/" + player_id + ".html?class=2;spanmax1=31+Jan+" + year +  ";spanval1=span;template=results;type=allround"
    html = urllib.urlopen(url, context=ctx).read()
    soup = bs(html)
    tables = soup.findAll("table")
    data = []
    for table in tables:
        if table.findParent("table") is None:
            #print table
            for tr in table.find_all('tr'):
                tds = tr.find_all('td')
                for t in tds:
                    data.append(t.text)
    try:
        if data[19] == 'Profile':
            ind = [22, 25, 26, 29, 30]
            odi = [float(data[i]) if data[i] != '-' else '-' for i in ind]
            #odi = [int(x) if x % 1 == 0 else x for x in odi]
        # table formatted differently
        elif data[19] == 'filtered':
            ind = [20, 23, 24, 27, 28]
            odi = [float(data[i]) if data[i] != '-' else '-' for i in ind]
            #odi = [int(x) if x % 1 == 0 else x for x in odi]
        # for debugging
        else:
            print "New Format"
            odi = []
    # no stats available
    except IndexError:
        odi = ['-','-','-','-','-']
    
    player = [player_id, year] 
    return player + odi, url
```

```
players = pd.read_csv('players.csv')
```

```
ids = [(str(int(players['player_id'][i])), str(players['year'][i])) for i in xrange(len(players))]
print ids
```

```
col_labels=['player_id','time_range','odis',
            'bat_avg','centuries','bowl_avg','fives']
odi_stats = pd.DataFrame(columns=col_labels)

for x in ids:
    x_stats = pre_data(x[0], x[1])
    print x_stats # for debugging
    row = pd.DataFrame([x_stats[0]], columns=col_labels)
    odi_stats = odi_stats.append(row)
    
```

```
new_cols = ['odis', 'bat_avg', 'centuries', 'bowl_avg', 'fives']
for c in new_cols:
    players[c] = ''
j=0
for i,r in odi_stats.iterrows():    
    for c in new_cols:
        players.at[j,c] = r[c]
    j+=1

print players
players.to_csv('odi_stats.csv')
```
